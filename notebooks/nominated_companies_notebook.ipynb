```python
import pandas as pd
import numpy as np
```

```python
tmdb_data_url = "TMDB_movie_dataset_v1.csv"
tmdb_df = pd.read_csv(tmdb_data_url)

company_nominations = {
    'A24': 17,
    'Netflix': 14,
    'Warner Bros': 12,
    'Walt Disney': 10,
    'Searchlight': 10,
    'Paramount': 9,
    'Universal': 8,
    'Focus Features': 7,
    'Neon': 4,
    'Orion Pictures/UAR': 2,
    'Sideshow': 2,
    'Sony Pictures Classics': 2
}


tmdb_df['Nominated'] = 0

for company, nominations in company_nominations.items():
    tmdb_df.loc[tmdb_df['production_companies'].str.contains(company, na=False, case=False), 'Nominated'] = 1


output_csv_path = "merged_nominated_company_data.csv"
tmdb_df.to_csv(output_csv_path, index=False)

print(tmdb_df.head())
```
